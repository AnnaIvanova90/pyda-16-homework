Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
movies = pd.read_csv('movies.csv')
movies.head(7)
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head(7)
```

```python
ratings_5 = ratings[ratings['rating'] == 5.0]
ratings_5.head(7)
```

```python
res_dict = {}
for item in ratings_5['movieId']:
    if item not in res_dict: 
        res_dict[item] = 1
    else: res_dict[item] += 1
```

```python
top_rank = max(res_dict.values())
for kino, rank in res_dict.items():
    if rank == top_rank: top_kino = kino

print(movies.loc[movies['movieId'] == top_kino])
```
